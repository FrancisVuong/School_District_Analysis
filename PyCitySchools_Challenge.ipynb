```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to change the path if needed.)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read the School Data and Student Data and store into a Pandas DataFrame
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Cleaning Student Names and Replacing Substrings in a Python String
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

# Iterate through the words in the "prefixes_suffixes" list and replace them with an empty space, "".
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"", regex= True)
#used regex = True to remove warning
# Check names.
student_data_df.head(10)
```

```python
# Merge Student_data_df with School_data to use for delieverable 2
## Create combined data frame named school_data_complete

school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
#check to see if it merged
school_data_complete_df.tail()
```

```python
#Setting Metrics for delieverable 2
d_student_count = school_data_complete_df.student_name.count()
d_school_count = len(school_data_complete_df.school_name.unique())
d_total_budget = school_data_df.budget.agg('sum')


d_avg_math = school_data_complete_df.math_score.agg('mean')
d_avg_reading = school_data_complete_df.reading_score.agg('mean')

passing_math_filter_df = school_data_complete_df[school_data_complete_df.math_score >= 70]
passing_reading_filter_df = school_data_complete_df[school_data_complete_df.reading_score >= 70]

passing_math_count = passing_math_filter_df.student_name.count()
passing_reading_count = passing_reading_filter_df.student_name.count()

d_passing_math_percent = passing_math_count/float(d_student_count)*100
d_passing_reading_percent = passing_reading_count/float(d_student_count)*100
d_overall_passing_percent = (d_passing_math_percent + d_passing_reading_percent)/2
```

```python
# Create a new data frame with outcomes 
district_summary_df = pd.DataFrame([{"Total Students": d_student_count, 
                                       "Total Schools": d_school_count, 
                                       "Total Budget": d_total_budget, 
                                      "Average Math Score": d_avg_math,
                                      "Average Reading Score": d_avg_reading,
                                      "% Passing Math": d_passing_math_percent,
                                      "% Passing Reading": d_passing_reading_percent,
                                      "% Overall Passing": d_overall_passing_percent}])
district_summary_df
```

```python
## Format new DataFrame
# Format the "Total Students" to have the comma for a thousands separator.
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
# Format the "Total Budget" to have the comma for a thousands separator, a decimal separator and a "$".
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
# Format the columns.
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.1f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.1f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.1f}".format)

# Display the data frame
district_summary_df
```
